```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = range(100)
Y = range(100)
plt.plot(X, Y)
plt.show()
```

```python
import numpy as np

X_1 = range(100)
Y_1 = [np.cos(value) for value in X_1]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X_2]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()
```

```python
import random

X_1 = [random.randint(1, 100) for _ in range(100)]
Y_1 = [np.cos(value) for value in X_1]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X_2]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()
```

### figure & axes
- matplotlib는 figure 안에 axes로 구성
- figure 위에 여러 개의 axes를 생성



```python
fig = plt.figure() # figure 반환
fig.set_size_inches(10, 5) # 크기 지정
ax_1 = fig.add_subplot(1, 2, 1) # 두 개의 plot 생성
ax_2 = fig.add_subplot(1, 2, 2)

ax_1.plot(X_1, Y_1, c='b')
ax_2.plot(X_2, Y_2, c='g')
plt.show() # show & flush
```

### subplot
- subplot의 순서를 grid로 작성



```python
X_1 = range(100)
Y_1 = [np.cos(value) for value in X_1]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X_2]

fig = plt.figure()
fig.set_size_inches(10, 5)
ax_1 = fig.add_subplot(1, 2, 1)
ax_2 = fig.add_subplot(1, 2, 2)

ax_1.plot(X_1, Y_1)
ax_2.plot(X_2, Y_2)
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10, 10)
# plt.style.use("ggplot")

ax = []
colors = ["b", "g", "r", "c", "m", "y", "k"]
for i in range(1, 7):
    ax.append(fig.add_subplot(2, 3, i))
    X_1 = np.arange(50)
    Y_1 = np.random.rand(50)
    c = colors[np.random.randint(1, len(colors))]

    ax[i - 1].plot(X_1, Y_1, c=c)

plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10, 5)
ax_1 = fig.add_subplot(1, 2, 1)
ax_2 = fig.add_subplot(1, 2, 2)

ax_1.plot(X_1, Y_1, c="b")
ax_2.plot(X_2, Y_2, c="g")
plt.show()  # show & flush
```

### set color
- color 속성을 사용
- float: 흑백, rgb color, predefined color 사용

### line graph
**월별 통화량 데이터 총합 구해보기**

```python
import pandas as pd

df = pd.read_csv("./phone_data.csv")
df.head()
```

```python
result = df.groupby(["month"])["duration"].sum()
result.plot()
```

```python
plt.plot(result.index, result, color="greenyellow", linestyle="dashed")
plt.title("Durations per month")
plt.show()
```

**일별 네트워크별 데이터의 총량을 다중 plot으로 작성하라**

```python
import datetime

def parsedate(date_info):
    return date_info[:9]
```

```python
df["day"] = df["date"].map(parsedate)
df.head()
```

```python
result = df.groupby(["network", "day"])["duration"].sum().reset_index()
result.head()
```

```python
result["network"].unique().tolist()
```

```python
result[result["network"] == "data"].head()
```

```python
fig = plt.figure()
fig.set_size_inches(10, 10)

network_types = result["network"].unique().tolist()
ax = []
for i in range(1, 7):
    ax.append(fig.add_subplot(2, 3, i))
    network_name = network_types[i - 1]

    plt.title(network_name)

    X_1 = result[result["network"] == network_name]["day"]
    Y_1 = result[result["network"] == network_name]["duration"]
    ax[i - 1].get_xaxis().set_visible(False)
    ax[i - 1].plot(X_1, Y_1)

plt.show()
```

### scatter
- marker: scatter 모양 지정

```python
data_1 = np.random.rand(512, 2)
data_2 = np.random.rand(512, 2)

plt.scatter(data_1[:, 0], data_1[:, 1], c="b", marker="x")
plt.scatter(data_2[:, 0], data_2[:, 1], c="r", marker="o")
plt.show()
```

```python
result = df.groupby(["network", "day"])["duration"].sum().reset_index()
networks = result["network"].unique().tolist()
colors = ["b", "g", "r", "c", "m", "y", "k"]

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

for network_name, color in zip(networks, colors):
    if network_name == "data":
        continue
    X_1 = result[result["network"] == network_name]["day"]
    Y_1 = result[result["network"] == network_name]["duration"]

    ax.scatter(X_1, Y_1, c=color, label=network_name)
    ax.get_xaxis().set_visible(False)

ax.set_xlabel("day")
ax.set_ylabel("duration")
plt.legend(shadow=True, fancybox=True, loc="upper right")
plt.show()
```

### bar chart

```python
data = [[5.0, 25.0, 50.0, 20.0], [4.0, 23.0, 51.0, 17], [6.0, 22.0, 52.0, 19]]

X = np.arange(0, 8, 2)

plt.bar(X + 0.00, data[0], color="b", width=0.50)
plt.bar(X + 0.50, data[1], color="g", width=0.50)
plt.bar(X + 1.0, data[2], color="r", width=0.50)
plt.xticks(X + 0.50, ("A", "B", "C", "D"))
plt.show()
```

```python
data = np.array([[5.0, 25.0, 50.0, 20.0], [4.0, 23.0, 51.0, 17], [6.0, 22.0, 52.0, 19]])

color_list = ["b", "g", "r"]
data_label = ["A", "B", "C"]
X = np.arange(data.shape[1])

for i in range(3):
    plt.bar(
        X,
        data[i],
        bottom=np.sum(data[:i], axis=0),
        color=color_list[i],
        label=data_label[i],
    )
plt.legend()
plt.show()
```

```python
A = [5.0, 30.0, 45.0, 22.0]
B = [5, 25, 50, 20]

X = range(4)

plt.bar(X, A, color="b")
plt.bar(X, B, color="r", bottom=60)
plt.show()
```

```python
women_pop = np.array([5, 30, 45, 22])
men_pop = np.array([5, 25, 50, 20])
X = np.arange(4)

plt.barh(X, women_pop, color="r")
plt.barh(X, -men_pop, color="b")
plt.show()
```

### seaborn

<img src="https://seaborn.pydata.org/_static/logo-wide-lightbg.svg" width=600>

- 기존 matplotlib에 기본 설정을 추가
- 복잡한 그래프를 간단하게 만들 수 있는 wrapper
- 간단한 코드 + 예쁜 결과
- matplotlib과 같은 기본적인 plot
- 손쉬운 설정으로 데이터 산출

#### predefined plots
- violinplot: boxplot에 distribution을 함께 표현
- stripplot: scatter와 category 정보를 함께 표현
- swarmplot: 분포와 함꼐 scatter를 함께 표현
- pointplot: category 별로 numeric의 평균, 신뢰구간 표시
- regplot: scatter + 선형 함수를 함께 표현

```python
import seaborn as sns

sns.set(style="darkgrid")
```

```python
tips = sns.load_dataset("tips")
fmri = sns.load_dataset("fmri")
```

```python
sns.set_style("whitegrid")
sns.lineplot(x="timepoint", y="signal", data=fmri)
```

```python
sns.lineplot(x="timepoint", y="signal", hue="event", data=fmri)
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
```

### multiple plot
- 한 개 이상의 도표를 하나의 plot에 작성
- axes를 사용해서 grid를 나누는 방법

#### predefined multiple plots
- replot: numeric 데이터 중심의 분포/선형 표시
- catplot: category 데이터 중심의 표시
- FacetGrid: 특정 조건에 따른 다양한 plot을 grid로 표시
- pairplot: 데이터 간의 상관관계 표시
- lmplot: regression 모델과 category 데이터를 함께 표시

```python
g = sns.FacetGrid(tips, col="time", row="sex")
```

```python
g = sns.FacetGrid(tips, col="time", row="sex")
g.map(sns.scatterplot, "total_bill", "tip")
```

```python
g = sns.FacetGrid(tips, col="time", row="sex")
g.map(sns.histplot, "total_bill")
```
