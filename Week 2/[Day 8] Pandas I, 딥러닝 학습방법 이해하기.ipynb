### lambda, map, apply

#### map function
- pandas의 series type의 데이터에도 map 함수 사용 가능
- function 대신 dict, sequence형 자료 등으로 대체 가능

```python
s1 = Series(np.arange(10))
s1.head()
```

```python
s1.map(lambda x: x**2).head()
```

```python
z = {1: 'A', 2: 'B', 3: 'C'}
s1.map(z).head() # dict type으로 데이터 교체, 없는 값은 NaN
```

```python
s2 = Series(np.arange(10, 20))
s1.map(s2).head() # 같은 위치의 데이터를 s2로 전환
```

```python
df = pd.read_csv("wages.csv")
df.head()
```

```python
df.sex.unique()
```

```python
df["sex_code"] = df.sex.map({"male": 0, "female": 1})
df.head()
```

#### replace function
- map 함수의 기능 중 데이터 변환 기능만 담당
- 데이터 변환 시 많이 사용하는 함수

```python
df.sex.replace({"male": 0, "female": 1}) # dict type 적용
df.head()
```

```python
df.sex.replace(["male", "female"], [0, 1], inplace=True) # df.replace([target list], [conversion list])
df.head() # inplace: 데이터 변환 결과를 적용
```

### apply function
- map과 달리, series 전체(column)에 해당 함수를 적용
- 입력 값이 series 데이터로 입력 받아 handling 가능

```python
df_info = df[["earn", "height", "age"]]
df_info.head()
```

```python
f = lambda x: x.max() - x.min()
df_info.apply(f) # 각 column별로 결과값 반환
```

- 내장 연산 함수를 사용할 때도 똑같은 효과를 거둘 수 있음
- mean, std 등 사용 가능

```python
df_info.sum()
```

```python
df_info.apply(sum)
```

- scalar 값 이외에 series 값의 반환도 가능함

```python
def f(x):
    return Series([x.min(), x.max()], index=["min", "max"])
df_info.apply(f)
```

#### applymap
- series 단위가 아닌 element 단위로 함수를 적용함
- series 단위에 apply를 적용시킬 때와 같은 효과

```python
f = lambda x: -x
df_info.applymap(f).head()
```

```python
f = lambda x: -x
df_info["earn"].apply(f).head()
```

#### unique
- series data의 유일한 값을 list로 반환함

```python
df.race.unique() # 유일한 race의 값 list
```

```python
np.array(dict(enumerate(df["race"].unique()))) # dict type으로
```

```python
# label index 값과 label 값 각각 추출
value = list(map(int, np.array(list(enumerate(df["race"].unique())))[:, 0].tolist()))
key = np.array(list(enumerate(df["race"].unique())), dtype=str)[:, 1].tolist()

value, key
```

```python
# label str -> index 값으로 변환
df["race"].replace(to_replace=key, value=value, inplace=True)
df.head()
```

```python
# sex에 대해서도 동일하게 적용
value = list(map(int, np.array(list(enumerate(df["sex"].unique())))[:, 0].tolist()))
key = np.array(list(enumerate(df["sex"].unique())), dtype=str)[:, 1].tolist()

value, key
```

```python
# "sex"와 "race" column의 index labelling
df["sex"].replace(to_replace=key, value=value, inplace=True)
df.head()
```
